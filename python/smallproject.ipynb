```python
app_names =[]
app_scores=[]
app_installs=[]
app_reviews=[]
app_rating=[]
i =0
while i<len(app_ids):
    d =app(app_ids[i],lang='en',country='in')
    app_names.append(d['title'])
    app_scores.append(d['score'])
    app_installs.append(d['installs'])
    app_reviews.append(d['reviews'])
    app_rating.append(d['ratings'])
    i+=1
    
```

```python
print(app_names)
```

```python
import pandas as pd
```

```python
df = pd.DataFrame({
    'App_name':app_names,
    'App_score':app_scores,
    'App_ratings':app_rating,
    'App_reviews': app_reviews,
    'App_installs':app_installs
})

    
```
